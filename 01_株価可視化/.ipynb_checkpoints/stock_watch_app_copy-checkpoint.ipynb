```python
#使用するモジュールのインポート
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

%matplotlib inline
```

```python
#株価の取得にはティッカーシンボルを使う
#ティッカーシンボル一覧：https://search.sbisec.co.jp/v2/popwin/info/stock/pop6040_usequity_list.html
#appleの株価
aapl = yf.Ticker('AAPL')
```

```python
#株価の表示(デフォルトでは１か月分表示)
aapl.history()
#引数で表示するデータ数を設定できる
days = 10
hist_aapl = aapl.history(period = f'{days}d')
hist_aapl
```

```python
#データ構造の確認
hist_aapl.index
hist_aapl.columns
hist_aapl.info()
```

```python
#indexに日付が割り当てられているのでリセットして番号を割り当てる
hist_aapl.reset_index()
```

```python
#データの整形
#インデックスのデータの形の確認
hist_aapl.index
```

```python
#データの形をstrftimeというメソッドで「日にち、月（英語表記）、年）」に変える
hist_aapl.index = hist_aapl.index.strftime('%d %B %Y')
hist_aapl
```

```python
#一日の終値のみ表示
hist_aapl = hist_aapl[['Close']]

#カラム名を会社名に
hist_aapl.columns = ['apple']
hist_aapl
```

```python
#終値の時間変化を見やすくするために転置
hist_aapl = hist_aapl.T
hist_aapl
```

```python
#インデックスにNameを追加
hist_aapl.index.name = 'Name'
```
