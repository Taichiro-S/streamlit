```python
#これまでの処理のまとめ

#yfinanceライブラリのインストール
!pip install yfinance

#使用するモジュールのインポート
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
%matplotlib inline

#appleの株価の取得
aapl = yf.Ticker('AAPL')

#指定した日にち分のデータ取得
days = 10
hist_aapl = aapl.history(period = f'{days}d')

#データの形をstrftimeというメソッドで「日にち、月（英語表記）、年）」に変える
hist_aapl.index = hist_aapl.index.strftime('%d %B %Y')

#一日の終値のみ表示
hist_aapl = hist_aapl[['Close']]

#カラム名を会社名に
hist_aapl.columns = ['apple']

#終値の時間変化を見やすくするために転置
hist_aapl = hist_aapl.T

#インデックスにNameを追加
hist_aapl.index.name = 'Name'
hist_aapl
```

```python
#複数社のデータを結合する関数
#日にちと会社名とTickerがペアになった辞書を引数とする
def get_data(days, tickers):
    df = pd.DataFrame()
    for company in tickers.keys():
        tkr = yf.Ticker(tickers[company])
        hist = tkr.history(period=f'{days}d')
        hist.index = hist.index.strftime('%d %B %Y')
        hist = hist[['Close']]
        hist.columns = [company]
        hist = hist.T
        hist.index.name = 'Name'
        df = pd.concat([df, hist])
    return df
```

```python
days = 20
tickers = {
    'apple': 'AAPL',
    'meta': 'META',
    'google': 'GOOGL',
    'microsoft': 'MSFT',
    'netflix': 'NFLX',
    'amazon': 'AMZN'
}

get_data(days, tickers)
```
